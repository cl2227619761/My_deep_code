```python
# 文件的输入输出
from skimage import io
```

```python
im = io.imread('./92091-1M.jpg')
```

```python
print('data type:', im.dtype)
```

```python
print('min value:', im.min())
print('max value:', im.max())
```

```python
# 浮点型的范围是[-1, 1]，skimage的一些图像处理例程处理的是浮点化的图像，所以可能导致输入是整型数组，输出是浮点数
from skimage import filters, color
```

```python
im_gray = color.rgb2gray(im)
```

```python
im_sobel = filters.sobel(im_gray)
```

```python
im_roberts = filters.roberts(im_gray)
im_scharr = filters.scharr(im_gray)
im_prewitt = filters.prewitt(im_gray)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
im_total = np.stack([im_sobel, im_roberts, im_scharr, im_prewitt])
```

```python
fig, axes = plt.subplots(ncols=2, nrows=2, figsize=(20, 20))
for i in range(2):
    for j in range(2):
        axes[i, j].imshow(im_total[j], cmap='gray')
        axes[i, j].axis('off')
plt.tight_layout()
plt.show()
```

```python
print('sobel im max:', im_sobel.max())
```

```python
im_hsv = color.rgb2hsv(im)
im_hsv.shape
```

```python
fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(10, 10))
axes[0].imshow(im)
axes[1].imshow(im_hsv)
axes[0].axis('off')
axes[1].axis('off')
plt.show()
```

```python
s_im = im_hsv[:, :, 1]
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 10))
axes[0].imshow(im)
axes[1].imshow(im_hsv)
axes[2].imshow(s_im, cmap='gray')
for i in range(3):
    axes[i].axis('off')
plt.show()
```

```python
s_im_prewitt = filters.prewitt(s_im)
h_im = im_hsv[:, :, 0]
s_im_total = np.stack([s_im, s_im_prewitt, im_gray, h_im])
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(10, 30))
for i in range(4):
    axes[i].imshow(s_im_total[i], cmap='gray')
    axes[i].axis('off')
plt.show()
```

```python
from skimage import exposure
```

```python
print('h channel data type:', h_im.dtype)
print('h channel min value:', h_im.min())
print('h channel max value:', h_im.max())
```

```python
s_hist, s_hist_centers = exposure.histogram(s_im, nbins=256)
```

```python
plt.plot(s_hist_centers, s_hist)
plt.show()
```

```python
plt.imshow(s_im > 0.6, cmap='gray')
plt.show()
```

```python
p2, p98 = np.percentile(s_im, (2, 98))
```

```python
print(p2, p98)
```

```python
gray_rescale = exposure.rescale_intensity(im_gray, in_range=(p2, p98))
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 10))
axes[0].imshow(im_gray, cmap='gray')
axes[1].imshow(gray_rescale, cmap='gray')
axes[0].axis('off')
axes[1].axis('off')
plt.show()
```
