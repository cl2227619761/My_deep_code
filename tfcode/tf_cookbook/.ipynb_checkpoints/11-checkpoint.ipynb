```python
# 多维图像的导入
import imageio
import numpy as np
```

```python
im = imageio.imread('../92091-1M.jpg')
```

```python
plt.imshow(im)
plt.show()
```

```python
import scipy.ndimage as ndi

print('data type:', im.dtype)
print('min value:', im.min())
print('max value:', im.max())
```

```python
hist = ndi.histogram(im, min=0, max=255, bins=256)
```

```python
cdf = hist.cumsum() / hist.sum()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)
axes[0].plot(hist, label='hist')
axes[1].plot(cdf, label='cumsum')
plt.show()
```

```python
im_equalized = cdf[im] * 255
```

```python
im_equalized.shape
```

```python
fig, axes = plt.subplots(ncols=1, nrows=2)
axes[0].imshow(im)
axes[1].imshow(im_equalized)
plt.show()
```

```python
mask = im > 230
```

```python
wbc = np.where(im < 230, im, 0)
```

```python
plt.imshow(wbc)
plt.axis('off')
plt.show()
```

```python
plt.imshow(wbc)
plt.show()
```

```python
import imageio
import matplotlib.pyplot as plt
import numpy as np
import scipy.ndimage as ndi
```

```python
im = imageio.imread('../2009107-7M.jpg')
```

```python
from skimage import color
```

```python
hsv_im = color.rgb2hsv(im)
```

```python
fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(20, 20))
axes[0].imshow(im)
axes[1].imshow(hsv_im)
axes[0].axis('off')
axes[1].axis('off')
plt.show()
```

```python
hsv_im.shape
```

```python
h_im = hsv_im[:, :, 0]
s_im = hsv_im[:, :, 1]
v_im = hsv_im[:, :, 2]
```

```python
fig, axe = plt.subplots(nrows=2, ncols=2, figsize=(20, 10), sharey=True)
axe[0, 0].imshow(im)
axe[0, 1].imshow(h_im)
axe[1, 0].imshow(s_im)
axe[1, 1].imshow(v_im)
for i in range(2):
    for j in range(2):
        axe[i, j].axis('off')
plt.show()
```

```python
print('s_im data type:', s_im.dtype)
print('s_im min value:', s_im.min())
print('s_im max value:', s_im.max())
```

```python
from skimage import img_as_ubyte
```

```python
s_im_int = img_as_ubyte(s_im)
```

```python
plt.imshow(s_im_int)
plt.show()
```

```python
print('s_im_int data type:', s_im_int.dtype)
print('s_im_int min value:', s_im_int.min())
print('s_im_int max value:', s_im_int.max())
```

```python
hist = ndi.histogram(s_im_int, min=0, max=255, bins=256)
plt.plot(hist)
plt.show()
```

```python
mask1 = np.where(s_im_int > 200, 1, 0)
plt.imshow(mask1, cmap='gray')
plt.show()
```

```python
tmp = np.stack([mask1] * 3, axis=-1) * im
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 20))
axes[0].imshow(im)
axes[1].imshow(tmp)
axes[0].axis('off')
axes[1].axis('off')
plt.show()
```
