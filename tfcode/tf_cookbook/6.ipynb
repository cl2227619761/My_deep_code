```python
# 损失函数
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

sess = tf.Session()
```

```python
# 回归的损失函数
x_vals = tf.linspace(-1., 1., 500)
target = tf.constant(0.)
```

```python
# L2 norm loss
l2_y_vals = tf.square(x_vals - target)
out = sess.run(l2_y_vals)
print(out)
```

```python
plt.plot(sess.run(x_vals), out, 'b', label='L2 norm loss')
plt.legend()
plt.show()
```

```python
help(tf.nn.l2_loss)
```

```python
res_l2 = tf.nn.l2_loss(x_vals - target)
print(sess.run(res_l2))
```

```python
# L1 norm loss，又称absolute loss function
l1_y_vals = tf.abs(x_vals - target)
l1_y_out = sess.run(l1_y_vals)
print(l1_y_out)
```

```python
plt.plot(sess.run(x_vals), l1_y_out, 'b', label='L1 norm loss')
plt.legend()
plt.show()
```

```python
# pseudo-huber loss
delta1 = tf.constant(0.25)
phuber_y_vals = tf.multiply(tf.square(delta1), tf.sqrt(1. + tf.square((target - x_vals)/delta1)) - 1.)
phuber_y_out = sess.run(phuber_y_vals)
print(phuber_y_out)
```

```python
delta2= tf.constant(5.)
phuber2_y_vals = tf.multiply(tf.square(delta2), tf.sqrt(1. + tf.square((target - x_vals)/delta2)) - 1.)
phuber2_y_out = sess.run(phuber2_y_vals)
print(phuber2_y_out)
```

```python
plt.plot(sess.run(x_vals), phuber_y_out, label='delta=0.25')
plt.plot(sess.run(x_vals), phuber2_y_out, label='delta=0.5')
plt.plot(sess.run(x_vals), out, label='l2 norm loss')
plt.plot(sess.run(x_vals), l1_y_out, label='l1 norm loss')
plt.legend()
plt.show()
```

```python
# classification loss functions
x_vals = tf.linspace(-3., 5, 500)
target = tf.constant(1.)
targets = tf.fill([500, ], 1.)
```

```python
print(sess.run(x_vals))
```

```python
print(sess.run(targets))
```

```python
# hinge loss
hinge_y_vals = tf.maximum(0., 1. - tf.multiply(target, x_vals))
hinge_y_out = sess.run(hinge_y_vals)
print(hinge_y_out)
```

```python
plt.plot(sess.run(x_vals), hinge_y_out, label='hinge loss')
plt.legend()
plt.show()
```

```python
# cross entropy loss, also referred to as the logistic loss function
x_vals = tf.linspace(0., 1., 500)
target = tf.constant(1.)

xentropy_y_vals = -tf.multiply(target, tf.log(x_vals)) - tf.multiply((1.-target), tf.log(1-x_vals))
xentropy_y_out = sess.run(xentropy_y_vals)

print(xentropy_y_out)
```

```python
plt.plot(sess.run(x_vals), xentropy_y_out, label='ce loss')
plt.legend()
plt.show()
```

```python
# sigmoid cross entropy loss，和cross entropy loss类似，只不过在计算损失之前，自动对x_vals进行了sigmoid转换
x_vals = tf.linspace(-3., 5., 500)

xentropy_sigmoid_y_vals = tf.nn.sigmoid_cross_entropy_with_logits(logits=x_vals, labels=targets)
xentropy_sigmoid_y_out = sess.run(xentropy_sigmoid_y_vals)
```

```python
print(xentropy_sigmoid_y_out)
```

```python
plt.plot(sess.run(x_vals), xentropy_sigmoid_y_out, label='ce sigmoid loss')
plt.legend()
plt.show()
```

```python
plt.plot(sess.run(x_vals), xentropy_y_out, label='ce loss')
plt.plot(sess.run(x_vals), xentropy_sigmoid_y_out, label='ce sigmoid loss')
plt.legend()
plt.show()
```

```python
plt.plot(sess.run(x_vals), xentropy_y_out, label='ce loss')
plt.plot(sess.run(x_vals), xentropy_sigmoid_y_out, label='ce sigmoid loss')
plt.plot(sess.run(x_vals), hinge_y_out, label='hinge loss')
plt.legend()
plt.show()
```

```python
# 加权cross-entropy loss
help(tf.nn.weighted_cross_entropy_with_logits)
```

```python
weight1 = tf.constant(0.5)
xentropy_weighted_y_vals = tf.nn.weighted_cross_entropy_with_logits(logits=x_vals, targets=targets, pos_weight=weight1)
xentropy_weighted_y_out = sess.run(xentropy_weighted_y_vals)
print(xentropy_weighted_y_out)
```

```python
plt.plot(sess.run(x_vals), xentropy_weighted_y_out, label='weighted ce loss')
plt.legend()
plt.show()
```

```python
plt.plot(sess.run(x_vals), xentropy_y_out, label='ce loss')
plt.plot(sess.run(x_vals), xentropy_sigmoid_y_out, label='ce sigmoid loss')
plt.plot(sess.run(x_vals), hinge_y_out, label='hinge loss')
plt.plot(sess.run(x_vals), xentropy_weighted_y_out, label='weighted ce loss')
plt.legend()
plt.show()
```

```python
weight2 = tf.constant(0.25)
xentropy_weighted2_y_vals = tf.nn.weighted_cross_entropy_with_logits(logits=x_vals, targets=targets, pos_weight=weight2)
xentropy_weighted2_y_out = sess.run(xentropy_weighted2_y_vals)
```

```python
plt.plot(sess.run(x_vals), xentropy_weighted2_y_out, label='weighted ce loss 0.25 x')
plt.plot(sess.run(x_vals), xentropy_weighted_y_out, label='weighted ce loss 0.5 x')
plt.legend()
plt.show()
```

```python
plt.plot(sess.run(x_vals - target), xentropy_weighted_y_out, label='weighted ce loss 0.5x')
plt.xlabel('distance')
plt.legend()
plt.show()
```
