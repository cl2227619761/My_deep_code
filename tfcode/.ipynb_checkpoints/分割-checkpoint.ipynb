```python
from PIL import Image, ImageTk
import os
from skimage import io, color, filters, data, segmentation, measure, morphology
from skimage.exposure import exposure
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
from skimage.morphology import disk
```

```python
import imageio

im = imageio.imread('./92091-1M.jpg')
plt.imshow(im)
plt.axis('off')
plt.show()
```

```python
img = io.imread('./92091-1M.jpg')
img_hsv = color.rgb2hsv(img)
s_channel = img_hsv[:, :, 1]
thresh = 0.7
dst =(s_channel >= thresh)*1.0 #根据阈值进行分割
#bw =morphology.closing(dst > thresh, morphology.square(8)) #闭运算
bw =morphology.opening(dst > thresh, morphology.square(16)) # 开运算

# bw=morphology.dilation(bw,morphology.square(2))
#bw=morphology.dilation(bw, disk(6))  # 先膨胀
#bw=morphology.erosion(bw, disk(4))  # 后腐蚀

cleared = bw.copy()  #复制
segmentation.clear_border(cleared)  #清除与边界相连的目标物
cleared=morphology.remove_small_objects(cleared,min_size=200,connectivity=1)

label_image =measure.label(cleared)  #连通区域标记
borders = np.logical_xor(bw, cleared) #异或
label_image[borders] = -1
image_label_overlay =color.label2rgb(label_image, image=dst) #不同标记用不同颜色显示

fig,(ax0,ax1)= plt.subplots(1,2, figsize=(20, 20))
ax0.imshow(cleared,plt.cm.gray)
ax1.imshow(image_label_overlay)
for region in measure.regionprops(label_image): #循环得到每一个连通区域属性集

    #忽略小区域
    if region.area < 200:
        continue

    #绘制外包矩形
    minr, minc, maxr, maxc = region.bbox
    rect = mpatches.Rectangle((minc, minr), maxc - minc, maxr - minr,
                              fill=False, edgecolor='red', linewidth=2)
    ax1.add_patch(rect)
fig.tight_layout()
plt.show()
```
